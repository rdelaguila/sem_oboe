```python
from ampligraph.evaluation import train_test_split_no_unseen 

num_test = int(len(X) * (20 / 100))

data = {}
X2 = df.to_numpy()
X = X2[:,[0,1,2,3]]

X = X.astype(str)
Xtrain = X[:,[0,1,2]]
positive_filter = Xtrain
data['train'], data['test'] = train_test_split_no_unseen(Xtrain, test_size=num_test, seed=0, allow_duplication=True) 
```

```python
data['train']
```

```python
from ampligraph.latent_features import ComplEx,TransE
```

```python
model = TransE(batches_count=100, 
                seed=0, 
                epochs=200, 
                k=150, 
                eta=5,
                optimizer='adam', 
                optimizer_params={'lr':1e-3},
                loss='multiclass_nll', 
                regularizer='LP', 
                regularizer_params={'p':3, 'lambda':1e-5}, 
                verbose=True)
```

```python
import tensorflow as tf
tf.logging.set_verbosity(tf.logging.ERROR)

model.fit(data['train'], early_stopping = False)
```

```python
save_model(model, './bbc_objects/best_model_kge_bbc_transe.pkl')
```

```python
model = restore_model('./bbc_objects/best_model_kge_bbc_transe.pkl')
```

```python
if model.is_fitted:
    print('The model is fit!')
else:
    print('The model is not fit! Did you skip a step?')
```

```python
from ampligraph.evaluation import evaluate_performance
positives_filter = Xtrain
ranks = evaluate_performance(data['test'], 
                             model=model, 
                             filter_triples=positives_filter,   # Corruption strategy filter defined above 
                             use_default_protocol=True, # corrupt subj and obj separately while evaluating
                             verbose=True)
```

```python
from ampligraph.evaluation import mr_score, mrr_score, hits_at_n_score

mrr = mrr_score(ranks)
print("MRR: %.2f" % (mrr))

hits_10 = hits_at_n_score(ranks, n=10)
print("Hits@10: %.2f" % (hits_10))
hits_3 = hits_at_n_score(ranks, n=3)
print("Hits@3: %.2f" % (hits_3))
hits_1 = hits_at_n_score(ranks, n=1)
print("Hits@1: %.2f" % (hits_1))
```

```python
def get_features_target_h2oFrame(mask):
    
    indices = list(mask.index)
    def get_embeddings(predicate):
      try:
        res= model.get_embeddings(predicate) #np.full(200, np.nan)
      except:
        res = np.full(300,np.nan)
      return res

    X = np.hstack((np.vstack(df.iloc[indices,:].subject.apply(get_embeddings).values),
                   #np.vstack(df.iloc[indices,:].relation.apply(get_embeddings).values),
                   np.vstack(df.iloc[indices,:].object.apply(get_embeddings).values)))
    y = df.iloc[indices,:].new_topic.values
     
    return X, y
```
